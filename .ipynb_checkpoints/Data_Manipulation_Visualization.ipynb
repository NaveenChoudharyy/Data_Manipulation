### 54. Write a Python program to get days between two dates. 

```python
start_date = x

while start_date is x:
    start_date_input = input('Please enter start date as dd/mm/yyyy: ')

    if not start_date_input:
        print('Invalid input. Please enter valid start date.')
    else:
        try:
            start_date = dt.datetime.strptime(start_date_input, '%d/%m/%Y')
        except ValueError:
            print('Invalid date format. Please enter start date in the format dd/mm/yyyy.')
            continue
```

```python
end_date = x

while end_date is x:
    end_date_input = input('Please enter end date as dd/mm/yyyy: ')

    if not end_date_input:
        print('Invalid input. Please enter valid end date.')
    else:
        try:
            end_date = dt.datetime.strptime(end_date_input, '%d/%m/%Y')
        except ValueError:
            print('Invalid date format. Please enter end date in the format dd/mm/yyyy.')
            continue
```

```python
def get_days_bet_2_days (start_date, end_date):
    return (end_date - start_date).days
```

```python
get_days_bet_2_days (start_date, end_date)
```

### 82. Create line chart for Total Revenue of all months with following properties
- X label name = Month
- Y label name = Total Revenue

```python
plt.figure( figsize=(10,4))
plt.ticklabel_format(style='plain')
sns.lineplot(x=company_data.Month, y=company_data.Total_Revenue, linewidth=2, color='Red', linestyle='--',marker='*', label='Monthly Revanue')
plt.title('Company Revenue over Months')
plt.xlabel('Months')
plt.ylabel('Revanue')
plt.grid(linestyle=':', color= 'green',alpha=.3)
plt.tight_layout()
plt.legend(loc=2)

plt.show()
```

### 83. Create line chart for Total Units of all months with following properties
- X label name = Month
- Y label name = Total Units
- Line Style dotted and Line-color should be red
- Show legend at the lower right location.

```python

plt.figure( figsize=(10,4))
sns.lineplot(x = company_data.Month, y = company_data.Total_Units, color='Red', linestyle='--',marker='o',label='Total Units for each month')


for i in range(len(company_data.Month)):
    plt.annotate(company_data.Total_Units[i], (company_data.Month[i], company_data.Total_Units[i]), textcoords="offset points", xytext=(0, 10), ha='center')




plt.title('Company Revenue over Months')
plt.xlabel('Months')
plt.ylabel('Total Units')
plt.grid(linestyle=':', color= 'white',alpha=.2,gapcolor='black')
plt.legend(loc=4)
plt.tight_layout()


plt.show()
```

### 84. Read all product sales data (Facecream, FaceWash, Toothpaste, Soap, Shampo, Moisturizer) and show it  using a multiline plot
- Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

```python
plt.figure(figsize=(10,5))


sns.lineplot(x=company_data.Month, y=company_data.FaceWash, color='black',label='Monthly units of facewash sold', marker='o')
sns.lineplot(x=company_data.Month, y=company_data.FaceCream, color='blue',label='Monthly units of FaceCream sold', marker='o')
sns.lineplot(x=company_data.Month, y=company_data.ToothPaste, color='brown',label='Monthly units of ToothPaste sold', marker='o')
sns.lineplot(x=company_data.Month, y=company_data.Soap, color='red',label='Monthly units of Soap sold', marker='o')
sns.lineplot(x=company_data.Month, y=company_data.Shampo, color='green',label='Monthly units of Shampo sold', marker='o')
sns.lineplot(x=company_data.Month, y=company_data.Moisturizer, color='#ce9b56',label='Monthly units of Moisturizer sold', marker='o')


plt.title('Monthly sales data of Facecream, FaceWash, Toothpaste, Soap, Shampo and Moisturizer')
plt.xlabel('Months')
plt.ylabel('Units sold')
plt.ylim(0,16000)


plt.grid(linestyle='-', color= 'lightgray',alpha=0.5)
plt.legend(loc=2,fontsize='small')
plt.tight_layout()
plt.show()
```

### 87. Create Histogram for Total Revenue

```python
plt.figure(figsize=(7,4))
plt.ticklabel_format(style='plain')

sns.histplot(x= company_data.Total_Revenue, bins=5,binrange=(company_data.Total_Revenue.min(),company_data.Total_Revenue.max()), color='red')

plt.title('Histogram of Total Revanue')
plt.xlabel('Total Revanue')
plt.ylabel('count of Revanue')

plt.grid(linestyle='--', color= 'green',alpha=0.2)
plt.tight_layout()
plt.show()
```

### 88. Calculate total sales data (quantity) for 2019 for each product and show it using a Pie chart. Understand percentage contribution from each product

```python
Pie_data = company_data.loc[:,['FaceCream','FaceWash','ToothPaste','Soap','Shampo','Moisturizer']].sum()
```

```python
plt.pie(Pie_data,labels=['FaceCream','FaceWash','ToothPaste','Soap','Shampo','Moisturizer'],labeldistance=1.05,colors=sns.color_palette('colorblind'), radius=1.7, autopct='%.1f%%',pctdistance=.8 )

plt.title('Total sales data (quantity) for year 2019 for each product',pad=70)


plt.tight_layout
plt.show()
```

### 89. Create line plots for Soap & Facewash of all months in a single plot using Subplot

```python
plt.figure(figsize=(15,4))


plt.subplot(1,2,1)
sns.lineplot(x=company_data.Month, y=company_data.Soap, color='r', markers='O')


for i in range(len(company_data.Month)):
    plt.annotate(company_data.Soap[i], (company_data.Month[i], company_data.Soap[i]), textcoords="offset points", xytext=(0, 10), ha='center')

plt.title('Monthly sales chart of Soap')
plt.xlabel('Months')
plt.ylabel('Units of Soap sold')
plt.grid()


plt.subplot(1,2,2)
sns.lineplot(x=company_data.Month, y=company_data.FaceWash, color='r')

for i in range(len(company_data.Month)):
    plt.annotate(company_data.FaceWash[i], (company_data.Month[i], company_data.FaceWash[i]), textcoords="offset points", xytext=(0, 10), ha='center')

plt.title('Monthly sales chart of FaceWash')
plt.xlabel('Months')
plt.ylabel('Units of FaceWash sold')
plt.grid()

plt.tight_layout()
plt.show()
```
